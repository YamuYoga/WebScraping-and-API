```python
import pandas as pd
import requests

def deg_to_text(deg): # cardinal directions for wind degrees
    return ["N","NNE","NE","ENE","E","ESE", "SE", "SSE","S","SSW","SW","WSW","W","WNW","NW","NNW"][round(deg/22.5)%16]

def getweather(cities= ["Berlin"], mode ='full'): # gets list of cities and mode
  url = "http://api.openweathermap.org/data/2.5/forecast?"
  API_key = "bb492ea5a591a1b1c090a6a952a139c4"
  fc_dict = {'city':[],'date':[],'weekday':[],'temperature [°C]':[], #empty dict
             'felt_temp':[],'pressure [hpa]':[],'description':[],
             'humidity [%]':[],'wind [m/s]':[], 'gusts':[],'direction':[],
             'clouds [%]':[],}
  for city in cities:
    try:
      query = f"q={city}&appid={API_key}&units=metric"
      weather_json = requests.get(url+query).json()
      fc_cnt = weather_json['cnt']
      for i in range(fc_cnt):
        fc_dict['city'].append(str.capitalize(city))
        fc_dict['date'].append(weather_json['list'][i]['dt_txt'])
        fc_dict['weekday'].append(weather_json['list'][i]['dt_txt'])
        fc_dict['temperature [°C]'].append(weather_json['list'][i]['main']['temp'])
        fc_dict['felt_temp'].append(weather_json['list'][i]['main']['feels_like'])
        fc_dict['pressure [hpa]'].append(weather_json['list'][i]['main']['pressure'])
        fc_dict['humidity [%]'].append(weather_json['list'][i]['main']['humidity'])
        fc_dict['description'].append(weather_json['list'][i]['weather'][0]['description'])
        fc_dict['wind [m/s]'].append(weather_json['list'][i]['wind']['speed'])
        fc_dict['gusts'].append(weather_json['list'][i]['wind']['gust'])
        fc_dict['direction'].append(deg_to_text(int(weather_json['list'][i]['wind']['deg'])))
        fc_dict['clouds [%]'].append(weather_json['list'][i]['clouds']['all'])
    except:
      print("An error occured.")
      raise
  fc_df = pd.DataFrame(fc_dict)
  fc_df['date'] = pd.to_datetime(fc_df['date'])
  fc_df['weekday'] = pd.to_datetime(fc_df['weekday']).dt.day_name()

  if mode == 'full':
    return fc_df
  elif mode == 'compact':
    fc_df['weekday'] = pd.to_datetime(fc_df['date']).dt.weekday
    compact_fc = (fc_df.groupby(['city','weekday'])
    .agg({'temperature [°C]':'mean',
         'humidity [%]':'mean','wind [m/s]':'mean','description':'first'})
    ).copy()
    # compact_fc['weekday'] = compact_fc['weekday'].dt.day_name()
    return compact_fc

city_list = ['Berlin','London']
getweather(city_list,"compact")
```
